```python
import pandas as pd

import warnings
warnings.simplefilter("ignore")

df = pd.read_excel("data-1.xlsx") # загрузка данных из файла

df.head() # отображение первых 5 строк исходных данных
```

```python
df = df.drop(columns = ["ID", "global_id", "ID_en", "MonthReport_en", "Calls_en"]) # удаление лишних столбцов
df = df.iloc[1: , :] # удаление первой строки, продублировала информацию о таблице

df.head()
```

```python
months = [
    ("Январь", "1"),
    ("Февраль", "2"),
    ("Март", "3"),
    ("Апрель", "4"),
    ("Май", "5"),
    ("Июнь", "6"),
    ("Июль", "7"),
    ("Август", "8"),
    ("Сентябрь", "9"),
    ("Октябрь", "10"),
    ("Ноябрь", "11"),
    ("Декабрь", "12"),
]

# функция замены названия месяца на его номер
def change(x):
    for month in months:
        if month[0] in x:
            return x.replace(month[0], month[1])
```

```python
df["MonthReport"] = df["MonthReport"].str.replace(" ", ".") # замена пробелов в строке с датой на точки
df["MonthReport"] = df["MonthReport"].apply(lambda x: "01." + change(x)) # замена исходных строк с датой на специальные формата ДД.ММ.ГГГГ
df["MonthReport"] = pd.to_datetime(df["MonthReport"]) # перевод строк с датой в формат даты

df.head()
```
